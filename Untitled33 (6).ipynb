```python
# ============================================================================
# CELL 6: Deep Learning Models - BERT for Emotion Classification
# ============================================================================
print("\n" + "="*70)
print("TRAINING BERT EMOTION CLASSIFIER")
print("="*70)

from transformers import (
    BertTokenizer, BertForSequenceClassification,
    AdamW, get_linear_schedule_with_warmup
)
from torch.utils.data import Dataset, DataLoader
import torch
from tqdm.auto import tqdm
import pandas as pd

# ------------------------------------------------------------
# 1ï¸âƒ£  Custom Dataset
# ------------------------------------------------------------
class EmotionDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=128):
        self.texts = list(texts)
        self.labels = list(labels)
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])
        enc = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding="max_length",
            truncation=True,
            return_attention_mask=True,
            return_tensors="pt"
        )
        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "label": torch.tensor(label, dtype=torch.long)
        }

# ------------------------------------------------------------
# 2ï¸âƒ£  Initialize Tokenizer & Model
# ------------------------------------------------------------
print("Initializing BERT model...")
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
num_emotions = len(train_df["emotion_label"].unique())

bert_model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=num_emotions
).to(device)

# ------------------------------------------------------------
# 3ï¸âƒ£  Create Datasets & Dataloaders
# ------------------------------------------------------------
train_dataset = EmotionDataset(
    texts=train_df["text_clean"].values,
    labels=train_df["emotion_label"].values,
    tokenizer=bert_tokenizer
)
val_dataset = EmotionDataset(
    texts=val_df["text_clean"].values,
    labels=val_df["emotion_label"].values,
    tokenizer=bert_tokenizer
)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader   = DataLoader(val_dataset,   batch_size=16, shuffle=False)

# ------------------------------------------------------------
# 4ï¸âƒ£  Optimizer & Scheduler
# ------------------------------------------------------------
optimizer = AdamW(bert_model.parameters(), lr=2e-5)
epochs = 3
total_steps = len(train_loader) * epochs

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=0,
    num_training_steps=total_steps
)

# ------------------------------------------------------------
# 5ï¸âƒ£  Training Loop
# ------------------------------------------------------------
best_val_acc = 0.0
training_stats = []

for epoch in range(epochs):
    print(f"\n{'='*70}\nEpoch {epoch+1}/{epochs}\n{'='*70}")

    # ---------- Training ----------
    bert_model.train()
    train_loss, train_correct, train_total = 0, 0, 0
    pbar = tqdm(train_loader, desc="Training")

    for batch in pbar:
        input_ids = batch["input_ids"].to(device)
        mask       = batch["attention_mask"].to(device)
        labels     = batch["label"].to(device)

        optimizer.zero_grad()
        outputs = bert_model(input_ids=input_ids, attention_mask=mask, labels=labels)
        loss, logits = outputs.loss, outputs.logits
        loss.backward()

        torch.nn.utils.clip_grad_norm_(bert_model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()

        train_loss += loss.item()
        preds = torch.argmax(logits, dim=1)
        train_correct += (preds == labels).sum().item()
        train_total   += labels.size(0)
        pbar.set_postfix({"loss": f"{loss.item():.4f}", "acc": f"{train_correct/train_total:.4f}"})

    avg_train_loss = train_loss / len(train_loader)
    train_acc = train_correct / train_total

    # ---------- Validation ----------
    bert_model.eval()
    val_loss, val_correct, val_total = 0, 0, 0
    with torch.no_grad():
        for batch in tqdm(val_loader, desc="Validation"):
            input_ids = batch["input_ids"].to(device)
            mask       = batch["attention_mask"].to(device)
            labels     = batch["label"].to(device)
            outputs = bert_model(input_ids=input_ids, attention_mask=mask, labels=labels)
            val_loss += outputs.loss.item()
            preds = torch.argmax(outputs.logits, dim=1)
            val_correct += (preds == labels).sum().item()
            val_total   += labels.size(0)

    avg_val_loss = val_loss / len(val_loader)
    val_acc = val_correct / val_total

    print(f"\nTrain Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val   Loss: {avg_val_loss:.4f} | Val   Acc: {val_acc:.4f}")

    training_stats.append({
        "epoch": epoch + 1,
        "train_loss": avg_train_loss,
        "train_acc": train_acc,
        "val_loss": avg_val_loss,
        "val_acc": val_acc
    })

    # Save best checkpoint
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        bert_model.save_pretrained("bert_emotion_classifier")
        bert_tokenizer.save_pretrained("bert_emotion_classifier")
        print("âœ“ Best model saved!")

print(f"\n{'='*70}\nBEST VALIDATION ACCURACY: {best_val_acc:.4f}\n{'='*70}")

pd.DataFrame(training_stats).to_csv("bert_training_stats.csv", index=False)
print("âœ“ Training stats saved to 'bert_training_stats.csv'")
```

```python
# ============================================================================
# CELL 7: LIWC-Style Linguistic Analysis
# ============================================================================
print("\n" + "="*70)
print("LIWC-STYLE LINGUISTIC ANALYSIS")
print("="*70)

from empath import Empath
import matplotlib.pyplot as plt
import pandas as pd
from tqdm.auto import tqdm

# -------------------------------------------------------------
# 1ï¸âƒ£  Initialize Empath lexicon
# -------------------------------------------------------------
lexicon = Empath()

EMOTION_CATEGORIES = [
    "positive_emotion", "negative_emotion", "anxiety", "anger",
    "sadness", "fear", "disgust", "joy", "anticipation", "surprise"
]
SOCIAL_CATEGORIES = [
    "social", "family", "friends", "communication", "help", "sympathy"
]

def analyze_text_liwc(text: str) -> dict:
    """Perform LIWC-style analysis using Empath categories."""
    if not isinstance(text, str):
        return {k: 0 for k in [
            "emotion_score", "positive_emotion", "negative_emotion",
            "anxiety", "sadness", "anger",
            "social_score", "help", "sympathy"
        ]}
    analysis = lexicon.analyze(text, normalize=True)
    return {
        "emotion_score": sum(analysis.get(cat, 0) for cat in EMOTION_CATEGORIES),
        "positive_emotion": analysis.get("positive_emotion", 0),
        "negative_emotion": analysis.get("negative_emotion", 0),
        "anxiety": analysis.get("anxiety", 0),
        "sadness": analysis.get("sadness", 0),
        "anger": analysis.get("anger", 0),
        "social_score": sum(analysis.get(cat, 0) for cat in SOCIAL_CATEGORIES),
        "help": analysis.get("help", 0),
        "sympathy": analysis.get("sympathy", 0)
    }

# -------------------------------------------------------------
# 2ï¸âƒ£  Analyze sample subset (for speed)
# -------------------------------------------------------------
SAMPLE_SIZE = 2000
print(f"\nAnalyzing {SAMPLE_SIZE} random contexts and responses ...")

context_analyses = [
    analyze_text_liwc(txt) for txt in tqdm(
        train_df["context"].sample(SAMPLE_SIZE, random_state=42),
        desc="Contexts"
    )
]
response_analyses = [
    analyze_text_liwc(txt) for txt in tqdm(
        train_df["utterance"].sample(SAMPLE_SIZE, random_state=42),
        desc="Responses"
    )
]

context_liwc_df  = pd.DataFrame(context_analyses)
response_liwc_df = pd.DataFrame(response_analyses)

# -------------------------------------------------------------
# 3ï¸âƒ£  Display descriptive summaries
# -------------------------------------------------------------
print("\nðŸ“Š LIWC Analysis Summary â€” Contexts:")
print(context_liwc_df[["emotion_score","positive_emotion","negative_emotion","social_score"]].describe())

print("\nðŸ“Š LIWC Analysis Summary â€” Responses:")
print(response_liwc_df[["emotion_score","positive_emotion","negative_emotion","social_score"]].describe())

# -------------------------------------------------------------
# 4ï¸âƒ£  Visualization
# -------------------------------------------------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# (a) Positive vs Negative emotions
emotion_data = pd.DataFrame({
    "Context": [context_liwc_df["positive_emotion"].mean(),
                context_liwc_df["negative_emotion"].mean()],
    "Response": [response_liwc_df["positive_emotion"].mean(),
                 response_liwc_df["negative_emotion"].mean()]
}, index=["Positive","Negative"])

emotion_data.plot(kind="bar", ax=axes[0,0], color=["#3498db","#e74c3c"])
axes[0,0].set_title("Positive vs Negative Emotions", fontweight="bold")
axes[0,0].set_ylabel("Average Score")
axes[0,0].set_xticklabels(["Positive","Negative"], rotation=0)

# (b) Specific emotions
specific_emotions = ["anxiety","sadness","anger"]
emotion_comparison = pd.DataFrame({
    "Context": [context_liwc_df[e].mean() for e in specific_emotions],
    "Response": [response_liwc_df[e].mean() for e in specific_emotions]
}, index=specific_emotions)

emotion_comparison.plot(kind="bar", ax=axes[0,1], color=["#9b59b6","#f39c12"])
axes[0,1].set_title("Specific Emotion Categories", fontweight="bold")
axes[0,1].set_ylabel("Average Score")

# (c) Social-score distributions
axes[1,0].hist(context_liwc_df["social_score"], bins=30, alpha=0.6,
               label="Context", color="#3498db")
axes[1,0].hist(response_liwc_df["social_score"], bins=30, alpha=0.6,
               label="Response", color="#e74c3c")
axes[1,0].set_title("Social Score Distribution", fontweight="bold")
axes[1,0].set_xlabel("Social Score")
axes[1,0].set_ylabel("Frequency")
axes[1,0].legend()

# (d) Help & Sympathy
help_sympathy = pd.DataFrame({
    "Context": [context_liwc_df["help"].mean(), context_liwc_df["sympathy"].mean()],
    "Response": [response_liwc_df["help"].mean(), response_liwc_df["sympathy"].mean()]
}, index=["Help","Sympathy"])

help_sympathy.plot(kind="bar", ax=axes[1,1], color=["#2ecc71","#e67e22"])
axes[1,1].set_title("Help & Sympathy Markers", fontweight="bold")
axes[1,1].set_ylabel("Average Score")

plt.tight_layout()
plt.savefig("liwc_analysis.png", dpi=300, bbox_inches="tight")
plt.show()

# -------------------------------------------------------------
# 5ï¸âƒ£  Save outputs
# -------------------------------------------------------------
context_liwc_df.to_csv("context_liwc_analysis.csv", index=False)
response_liwc_df.to_csv("response_liwc_analysis.csv", index=False)
print("\nâœ“ LIWC analysis complete and saved!")
```

```python
# ============================================================================
# CELL 10: Build Emotional Support Chatbot (LLAMA3-8B via HTTP API - NO SENTINEL)
# ============================================================================
print("\n" + "="*70)
print("BUILDING EMOTIONAL SUPPORT CHATBOT (LLAMA3-8B, NO SENTINEL, NO OLLAMA LIB)")
print("="*70)

import requests
from datetime import datetime
from collections import Counter
import numpy as np
import torch
from nltk.sentiment import SentimentIntensityAnalyzer
from empath import Empath


# =========================================================
# FUNCTION TO CALL OLLAMA WITHOUT THE PYTHON PACKAGE
# =========================================================
def llama3_generate(prompt):
    """Call Ollama locally via REST API"""
    url = "http://localhost:11434/api/generate"
    payload = {
        "model": "llama3:8b",
        "prompt": prompt,
        "stream": False
    }

    try:
        response = requests.post(url, json=payload)
        response = response.json()
        return response.get("response", "")
    except Exception as e:
        return f"[ERROR contacting Ollama: {e}]"


print("âœ“ Using LLaMA3-8B via local HTTP API (no pydantic, no sentinel)\n")


# =========================================================
# CHATBOT CLASS
# =========================================================
class EmotionalSupportChatbot:
    """Emotional chatbot using BERT + Empath + VADER + LLaMA3 (HTTP API)"""

    def __init__(self, emotion_classifier, tokenizer_classifier, id2emotion, device):
        self.emotion_classifier = emotion_classifier
        self.tokenizer_classifier = tokenizer_classifier
        self.id2emotion = id2emotion
        self.device = device

        self.conversation_history = []
        self.sia = SentimentIntensityAnalyzer()
        self.lexicon = Empath()

    # ----------------------------------------------------------
    # EMOTION DETECTION
    # ----------------------------------------------------------
    def detect_emotion(self, text):
        self.emotion_classifier.eval()

        enc = self.tokenizer_classifier.encode_plus(
            text,
            add_special_tokens=True,
            max_length=128,
            truncation=True,
            padding="max_length",
            return_attention_mask=True,
            return_tensors="pt"
        )

        input_ids = enc["input_ids"].to(self.device)
        attention_mask = enc["attention_mask"].to(self.device)

        with torch.no_grad():
            out = self.emotion_classifier(input_ids, attention_mask=attention_mask)
            probs = torch.softmax(out.logits, dim=1)
            top_probs, top_idx = torch.topk(probs, k=3)

        emotions = []
        for p, i in zip(top_probs[0], top_idx[0]):
            emotions.append({
                "emotion": self.id2emotion.get(i.item(), "neutral"),
                "confidence": float(p.item())
            })
        return emotions

    # ----------------------------------------------------------
    # SENTIMENT ANALYSIS
    # ----------------------------------------------------------
    def analyze_sentiment(self, text):
        return self.sia.polarity_scores(text)

    # ----------------------------------------------------------
    # EMPATHY ANALYSIS
    # ----------------------------------------------------------
    def analyze_empathy(self, text):
        analysis = self.lexicon.analyze(text, normalize=True)
        keys = ["positive_emotion", "affection", "sympathy", "help", "social", "communication"]
        return {"empathy_score": sum([analysis.get(k, 0) for k in keys])}

    # ----------------------------------------------------------
    # RESPONSE GENERATION (Ollama REST API)
    # ----------------------------------------------------------
    def generate_response(self, user_input, detected):
        emotion = detected[0]["emotion"]

        empathy_prefix = {
            "sad": "I'm really sorry you're feeling this way. ",
            "anxious": "I can sense your anxiety. ",
            "angry": "I hear your frustration. ",
            "lonely": "You're not alone. I'm here with you. ",
            "afraid": "It's okay to feel scared. ",
            "excited": "That's wonderful! ",
            "happy": "I'm glad you're feeling good! "
        }

        prefix = empathy_prefix.get(emotion, "I'm here to support you. ")

        # previous conversation context
        ctx = ""
        if len(self.conversation_history) > 0:
            ctx = "\n".join([f"User: {c['user']}\nBot: {c['bot']}" 
                             for c in self.conversation_history[-3:]])

        prompt = f"""
You are an empathetic emotional support AI.

Conversation so far:
{ctx}

Detected emotion: {emotion}
User message: {user_input}

Start your reply with: "{prefix}"
And continue with emotional understanding, validation, and empathy.
"""

        return llama3_generate(prompt).strip()

    # ----------------------------------------------------------
    # MAIN CHAT LOGIC
    # ----------------------------------------------------------
    def chat(self, user_input):

        emotions = self.detect_emotion(user_input)
        sentiment = self.analyze_sentiment(user_input)
        response = self.generate_response(user_input, emotions)

        empathy = self.analyze_empathy(response)
        resp_sent = self.analyze_sentiment(response)

        interaction = {
            "timestamp": datetime.now().isoformat(),
            "user": user_input,
            "bot": response,
            "detected_emotion": emotions[0]["emotion"],
            "emotion_confidence": emotions[0]["confidence"],
            "user_sentiment": sentiment["compound"],
            "response_empathy_score": empathy["empathy_score"],
            "response_sentiment": resp_sent["compound"]
        }

        self.conversation_history.append(interaction)
        return response, interaction


# =========================================================
# INITIALIZATION
# =========================================================
print("Initializing chatbot...\n")

our_chatbot = EmotionalSupportChatbot(
    emotion_classifier=bert_model,
    tokenizer_classifier=bert_tokenizer,
    id2emotion=id2emotion,
    device=device
)

print("âœ“ Emotional Support Chatbot READY (LLaMA3-8B, No Sentinel!)")
```

```python
# ============================================================================
# CELL 11: Create Baseline Chatbot for Comparison (Updated + Matched Structure)
# ============================================================================
print("\n" + "="*70)
print("CREATING BASELINE CHATBOT FOR COMPARISON (UPDATED)")
print("="*70)

from nltk.sentiment import SentimentIntensityAnalyzer
from empath import Empath
from datetime import datetime
import numpy as np
from collections import Counter


class BaselineChatbot:
    """Simple rule-based baseline for comparison with the advanced model"""

    def __init__(self):
        self.sia = SentimentIntensityAnalyzer()
        self.lexicon = Empath()
        self.conversation_history = []

        # Keyword-based emotion detection
        self.emotion_keywords = {
            'sad': ['sad', 'unhappy', 'depressed', 'down', 'miserable'],
            'happy': ['happy', 'joyful', 'excited', 'great', 'wonderful'],
            'angry': ['angry', 'mad', 'furious', 'annoyed', 'irritated'],
            'anxious': ['anxious', 'worried', 'nervous', 'stressed', 'tense'],
            'afraid': ['afraid', 'scared', 'frightened', 'terrified', 'fear'],
            'lonely': ['lonely', 'alone', 'isolated', 'abandoned']
        }

        # Basic responses (low empathy)
        self.responses = {
            'sad': "I'm sorry to hear that. I hope things improve soon.",
            'happy': "That's nice to hear!",
            'angry': "I understand you're upset.",
            'anxious': "Try to take a deep breath. It will be okay.",
            'afraid': "Try not to worry too much.",
            'lonely': "You're not really alone.",
            'neutral': "I see. Tell me more."
        }

    # --------------------------------------------------------------
    # SIMPLE EMOTION DETECTION (keywords)
    # --------------------------------------------------------------
    def detect_emotion(self, text):
        text_lower = text.lower()

        for emotion, keywords in self.emotion_keywords.items():
            for keyword in keywords:
                if keyword in text_lower:
                    return emotion, 0.55  # fixed lower confidence

        return 'neutral', 0.40

    # --------------------------------------------------------------
    # EMPATHY SCORE (using Empath)
    # --------------------------------------------------------------
    def analyze_empathy(self, text):
        analysis = self.lexicon.analyze(text, normalize=True)
        keys = ["positive_emotion", "affection", "sympathy", "help", "social", "communication"]
        return sum([analysis.get(k, 0) for k in keys])

    # --------------------------------------------------------------
    # RESPONSE GENERATION
    # --------------------------------------------------------------
    def chat(self, user_input):

        if not user_input.strip():
            return "Hello, how can I help you?", None

        # Detect emotion (keyword-based)
        emotion, confidence = self.detect_emotion(user_input)

        # User sentiment
        sentiment = self.sia.polarity_scores(user_input)

        # Choose simple response
        response = self.responses.get(emotion, self.responses['neutral'])

        # Empathy score (low because rules are simple)
        empathy_score = self.analyze_empathy(response)

        # Response sentiment
        resp_sent = self.sia.polarity_scores(response)

        # Store interaction
        interaction = {
            'timestamp': datetime.now().isoformat(),
            'user': user_input,
            'bot': response,
            'detected_emotion': emotion,
            'emotion_confidence': confidence,
            'user_sentiment': sentiment['compound'],
            'response_empathy_score': empathy_score,
            'response_sentiment': resp_sent['compound']
        }

        self.conversation_history.append(interaction)

        return response, interaction

    # --------------------------------------------------------------
    # SUMMARY
    # --------------------------------------------------------------
    def get_summary(self):
        if not self.conversation_history:
            return None

        emotions = [h['detected_emotion'] for h in self.conversation_history]

        return {
            'total_interactions': len(self.conversation_history),
            'emotions_detected': dict(Counter(emotions)),
            'avg_emotion_confidence': float(np.mean([h['emotion_confidence']
                                                     for h in self.conversation_history])),
            'avg_user_sentiment': float(np.mean([h['user_sentiment']
                                                 for h in self.conversation_history])),
            'avg_response_empathy': float(np.mean([h['response_empathy_score']
                                                   for h in self.conversation_history])),
            'avg_response_sentiment': float(np.mean([h['response_sentiment']
                                                     for h in self.conversation_history]))
        }


# Initialize baseline
baseline_chatbot = BaselineChatbot()
print("âœ“ Baseline chatbot created for comparison (updated)")
```

```python
# ============================================================================
# CELL 10: Build Complete Emotional Support Chatbot (LLAMA3-8B, OLLAMA)
# ============================================================================
print("\n" + "="*70)
print("BUILDING EMOTIONAL SUPPORT CHATBOT (LLAMA3-8B)")
print("="*70)

import ollama
from datetime import datetime
from collections import Counter
import numpy as np
import torch
from nltk.sentiment import SentimentIntensityAnalyzer
from empath import Empath

print("âœ“ Using OLLAMA LLaMA3:8B model for response generation\n")


# ============================================================================
# CHATBOT CLASS
# ============================================================================
class EmotionalSupportChatbot:
    """Emotional Support Chatbot using:
       - BERT Emotion Classifier
       - Empath Lexicon
       - VADER Sentiment
       - OLLAMA LLaMA3-8B for generation
    """

    def __init__(self, emotion_classifier, tokenizer_classifier, id2emotion, device):
        self.emotion_classifier = emotion_classifier
        self.tokenizer_classifier = tokenizer_classifier
        self.id2emotion = id2emotion
        self.device = device

        self.conversation_history = []
        self.sia = SentimentIntensityAnalyzer()
        self.lexicon = Empath()

    # ----------------------------------------------------------------------
    # EMOTION DETECTION
    # ----------------------------------------------------------------------
    def detect_emotion(self, text):
        self.emotion_classifier.eval()

        encoded = self.tokenizer_classifier.encode_plus(
            text,
            add_special_tokens=True,
            max_length=128,
            truncation=True,
            padding="max_length",
            return_attention_mask=True,
            return_tensors="pt"
        )

        input_ids = encoded["input_ids"].to(self.device)
        attention_mask = encoded["attention_mask"].to(self.device)

        with torch.no_grad():
            output = self.emotion_classifier(input_ids, attention_mask=attention_mask)
            probs = torch.softmax(output.logits, dim=1)
            top_probs, top_idx = torch.topk(probs, k=3)

        emotions = []
        for p, i in zip(top_probs[0], top_idx[0]):
            emotions.append({
                "emotion": self.id2emotion.get(i.item(), "neutral"),
                "confidence": float(p.item())
            })

        return emotions

    # ----------------------------------------------------------------------
    # SENTIMENT
    # ----------------------------------------------------------------------
    def analyze_sentiment(self, text):
        return self.sia.polarity_scores(text)

    # ----------------------------------------------------------------------
    # EMPATHY MARKERS (Empath library)
    # ----------------------------------------------------------------------
    def analyze_empathy(self, text):
        analysis = self.lexicon.analyze(text, normalize=True)
        keys = ["positive_emotion", "affection", "sympathy", "help", "social", "communication"]
        score = sum([analysis.get(k, 0) for k in keys])
        return {"empathy_score": score}

    # ----------------------------------------------------------------------
    # GENERATE RESPONSE USING OLLAMA LLAMA3-8B
    # ----------------------------------------------------------------------
    def generate_response(self, user_input, detected):

        user_emotion = detected[0]["emotion"]

        empathy_prefix = {
            "sad": "I'm really sorry you're feeling this way. ",
            "anxious": "I can sense your anxiety. ",
            "angry": "I hear your frustration. ",
            "lonely": "You're not alone. I'm here with you. ",
            "afraid": "It's okay to feel scared. ",
            "excited": "That's exciting! ",
            "happy": "I'm glad to hear that! "
        }

        prefix_text = empathy_prefix.get(user_emotion, "I'm here to support you. ")

        # Last 3 conversation turns
        conversation_context = ""
        if len(self.conversation_history) > 0:
            conversation_context = "\n".join([
                f"User: {entry['user']}\nBot: {entry['bot']}"
                for entry in self.conversation_history[-3:]
            ])

        # MASTER PROMPT
        final_prompt = f"""
You are an extremely empathetic emotional-support AI. Respond in a warm, caring, human-like tone. 
Avoid robotic or generic responses.

Conversation context:
{conversation_context}

Detected emotion: {user_emotion}
User message: {user_input}

Start your message with: "{prefix_text}"
Then continue with emotional understanding, validation, and gentle support.
"""

        # OLLAMA CALL
        result = ollama.generate(
            model="llama3:8b",
            prompt=final_prompt
        )

        bot_response = result["response"].strip()
        return bot_response

    # ----------------------------------------------------------------------
    # MAIN CHAT FUNCTION
    # ----------------------------------------------------------------------
    def chat(self, user_input):

        emotions = self.detect_emotion(user_input)
        sentiment = self.analyze_sentiment(user_input)
        response = self.generate_response(user_input, emotions)

        empathy_metrics = self.analyze_empathy(response)
        response_sentiment = self.analyze_sentiment(response)

        interaction = {
            "timestamp": datetime.now().isoformat(),
            "user": user_input,
            "bot": response,
            "detected_emotion": emotions[0]["emotion"],
            "emotion_confidence": emotions[0]["confidence"],
            "user_sentiment": sentiment["compound"],
            "response_empathy_score": empathy_metrics["empathy_score"],
            "response_sentiment": response_sentiment["compound"]
        }

        self.conversation_history.append(interaction)
        return response, interaction

    # ----------------------------------------------------------------------
    # SUMMARY
    # ----------------------------------------------------------------------
    def get_summary(self):
        if not self.conversation_history:
            return None

        return {
            "total_interactions": len(self.conversation_history),
            "emotions_detected": dict(Counter([x["detected_emotion"] for x in self.conversation_history])),
            "avg_emotion_confidence": float(np.mean([x["emotion_confidence"] for x in self.conversation_history])),
            "avg_user_sentiment": float(np.mean([x["user_sentiment"] for x in self.conversation_history])),
            "avg_response_empathy": float(np.mean([x["response_empathy_score"] for x in self.conversation_history])),
            "avg_response_sentiment": float(np.mean([x["response_sentiment"] for x in self.conversation_history]))
        }


# ============================================================================
# INITIALIZE CHATBOT
# ============================================================================
print("Initializing Emotional Support Chatbot...\n")

our_chatbot = EmotionalSupportChatbot(
    emotion_classifier=bert_model,
    tokenizer_classifier=bert_tokenizer,
    id2emotion=id2emotion,
    device=device
)

print("âœ“ Emotional Support Chatbot (LLaMA3-8B) initialized successfully!")
```

```python

# ============================================================================
# CELL 11: Baseline Chatbot for Comparison (Compatible With OLLAMA Chatbot)
# ============================================================================
print("\n" + "="*70)
print("CREATING BASELINE CHATBOT FOR COMPARISON (Ollama-Compatible)")
print("="*70)

from nltk.sentiment import SentimentIntensityAnalyzer
from datetime import datetime
from collections import Counter
import numpy as np
import json

# ---------------------------------------------------------
# BASELINE RULE-BASED CHATBOT
# ---------------------------------------------------------
class BaselineChatbot:
    """Very simple rule-based baseline chatbot for evaluation."""

    def __init__(self):
        self.sia = SentimentIntensityAnalyzer()
        self.conversation_history = []

        # Keyword triggers for emotions
        self.emotion_keywords = {
            "sad": ["sad", "down", "depressed", "unhappy", "cry"],
            "happy": ["happy", "joy", "excited", "great", "wonderful"],
            "angry": ["angry", "mad", "furious", "annoyed", "irritated"],
            "anxious": ["anxious", "worried", "nervous", "stressed", "panic"],
            "afraid": ["afraid", "scared", "terrified", "fear"],
            "lonely": ["lonely", "alone", "isolated", "abandoned"]
        }

        # Simple canned responses (low empathy baseline)
        self.responses = {
            "sad": "I'm sorry you're feeling this way. I hope things improve soon.",
            "happy": "That's wonderful! I'm really happy for you.",
            "angry": "I understand you're frustrated. Try to calm down slowly.",
            "anxious": "That sounds stressful. Try taking a deep breath.",
            "afraid": "It's okay to feel scared sometimes. Youâ€™re stronger than you think.",
            "lonely": "You're not alone. I'm here with you.",
            "neutral": "I see. Tell me more about that."
        }

    # ---------------------------------------------------------
    def detect_emotion(self, text):
        """Very simple keyword-based emotion detection."""
        if not isinstance(text, str) or text.strip() == "":
            return "neutral", 0.5

        t = text.lower()
        for emo, kw_list in self.emotion_keywords.items():
            for kw in kw_list:
                if kw in t:
                    return emo, 0.7
        return "neutral", 0.5

    # ---------------------------------------------------------
    def chat(self, user_input):
        """One chat turn for the baseline model."""
        if not isinstance(user_input, str) or user_input.strip() == "":
            user_input = ""

        emotion, conf = self.detect_emotion(user_input)
        sentiment = self.sia.polarity_scores(user_input)
        response = self.responses.get(emotion, self.responses["neutral"])

        meta = {
            "timestamp": datetime.now().isoformat(),
            "user": user_input,
            "bot": response,
            "detected_emotion": emotion,
            "emotion_confidence": conf,
            "user_sentiment": sentiment.get("compound", 0.0),
            "response_empathy_score": 0.2,   # fixed low empathy
            "response_sentiment": 0.3        # neutral-ish response
        }

        self.conversation_history.append(meta)
        return response, meta

    # ---------------------------------------------------------
    def get_summary(self):
        """Return baseline's evaluation metrics."""
        if not self.conversation_history:
            return None

        hist = self.conversation_history

        return {
            "total_interactions": len(hist),
            "emotions_detected": dict(Counter([h["detected_emotion"] for h in hist])),
            "avg_emotion_confidence": float(np.mean([h["emotion_confidence"] for h in hist])),
            "avg_user_sentiment": float(np.mean([h["user_sentiment"] for h in hist])),
            "avg_response_empathy": float(np.mean([h["response_empathy_score"] for h in hist])),
            "avg_response_sentiment": float(np.mean([h["response_sentiment"] for h in hist])),
        }

    # ---------------------------------------------------------
    def save_conversation(self, filename="baseline_conversation.json"):
        with open(filename, "w") as f:
            json.dump(self.conversation_history, f, indent=2)


# ---------------------------------------------------------
# Instantiate baseline chatbot
# ---------------------------------------------------------
baseline_chatbot = BaselineChatbot()
print("âœ“ Baseline chatbot created successfully.")


# ---------------------------------------------------------
# COMPARISON WRAPPER (For Cell 12)
# ---------------------------------------------------------
def compare_chatbots(text_samples, baseline, our_bot):
    """
    Compare Baseline Chatbot vs Our Emotional Support Chatbot (Ollama version).
    - Produces logs for both bots
    - Returns summaries + logs for Cell 12 visualization
    """

    baseline_logs = []
    our_logs = []

    for text in text_samples:

        # Baseline
        _, b_meta = baseline.chat(text)
        baseline_logs.append(b_meta)

        # Our emotional chatbot (Ollama response)
        resp, o_meta = our_bot.chat(text)

        # Sometimes our_bot may return None (error case)
        if o_meta is None:
            o_meta = {
                "timestamp": datetime.now().isoformat(),
                "user": text,
                "bot": resp,
                "detected_emotion": "neutral",
                "emotion_confidence": 0.0,
                "user_sentiment": 0.0,
                "response_empathy_score": 0.0,
                "response_sentiment": 0.0
            }

        our_logs.append(o_meta)

    return {
        "baseline": {
            "summary": baseline.get_summary(),
            "logs": baseline_logs
        },
        "our_bot": {
            "summary": our_bot.get_summary(),
            "logs": our_logs
        }
    }


# ---------------------------------------------------------
# Test Comparison (optional)
# ---------------------------------------------------------
example_inputs = [
    "I feel so sad today.",
    "I'm extremely anxious about my future.",
    "I'm very happy with my results!"
]

results = compare_chatbots(example_inputs, baseline_chatbot, our_chatbot)
print("âœ“ Comparison helper ready.")
print(results)
```

```python

# ============================================================================
# CELL 11: Create Baseline Chatbot for Comparison
# ============================================================================
print("\n" + "="*70)
print("CREATING BASELINE CHATBOT FOR COMPARISON")
print("="*70)

class BaselineChatbot:
    """Simple rule-based chatbot for comparison"""
    
    def __init__(self):
        self.sia = SentimentIntensityAnalyzer()
        self.conversation_history = []
        
        # Simple keyword-based emotion detection
        self.emotion_keywords = {
            'sad': ['sad', 'unhappy', 'depressed', 'down', 'miserable'],
            'happy': ['happy', 'joyful', 'excited', 'great', 'wonderful'],
            'angry': ['angry', 'mad', 'furious', 'annoyed', 'irritated'],
            'anxious': ['anxious', 'worried', 'nervous', 'stressed', 'tense'],
            'afraid': ['afraid', 'scared', 'frightened', 'terrified', 'fear'],
            'lonely': ['lonely', 'alone', 'isolated', 'abandoned']
        }
        
        # Generic responses
        self.responses = {
            'sad': "I'm sorry to hear that. Things will get better.",
            'happy': "That's great! I'm happy for you.",
            'angry': "I understand you're upset. Try to calm down.",
            'anxious': "Don't worry too much. It will be okay.",
            'afraid': "There's nothing to be afraid of. Stay strong.",
            'lonely': "You're not alone. Things will improve.",
            'neutral': "I see. Tell me more about how you feel."
        }
    
    def detect_emotion(self, text):
        """Simple keyword-based emotion detection"""
        text_lower = text.lower()
        
        for emotion, keywords in self.emotion_keywords.items():
            for keyword in keywords:
                if keyword in text_lower:
                    return emotion, 0.7  # Fixed confidence
        
        return 'neutral', 0.5
    
    def chat(self, user_input):
        """Generate response"""
        if not user_input.strip():
            return "How can I help you today?"
        
        # Detect emotion
        emotion, confidence = self.detect_emotion(user_input)
        
        # Get sentiment
        sentiment = self.sia.polarity_scores(user_input)
        
        # Generate response
        response = self.responses.get(emotion, self.responses['neutral'])
        
        # Store interaction
        interaction = {
            'timestamp': datetime.now().isoformat(),
            'user': user_input,
            'bot': response,
            'detected_emotion': emotion,
            'emotion_confidence': confidence,
            'user_sentiment': sentiment['compound'],
            'response_empathy_score': 0.2,  # Low fixed empathy
            'response_sentiment': 0.3  # Neutral sentiment
        }
        
        self.conversation_history.append(interaction)
        
        return response, interaction
    
    def get_summary(self):
        """Get summary"""
        if not self.conversation_history:
            return None
        
        emotions = [h['detected_emotion'] for h in self.conversation_history]
        
        return {
            'total_interactions': len(self.conversation_history),
            'emotions_detected': dict(Counter(emotions)),
            'avg_emotion_confidence': np.mean([h['emotion_confidence'] 
                                              for h in self.conversation_history]),
            'avg_user_sentiment': np.mean([h['user_sentiment'] 
                                          for h in self.conversation_history]),
            'avg_response_empathy': np.mean([h['response_empathy_score'] 
                                            for h in self.conversation_history]),
            'avg_response_sentiment': np.mean([h['response_sentiment'] 
                                              for h in self.conversation_history])
        }

# Initialize baseline
baseline_chatbot = BaselineChatbot()
print("âœ“ Baseline chatbot created for comparison")
```

```python

# ============================================================================
# CELL 13: Advanced Evaluation Metrics (BLEU & ROUGE Compatible with OLLAMA)
# ============================================================================
print("\n" + "="*70)
print("ADVANCED EVALUATION METRICS (Ollama-Compatible)")
print("="*70)

import numpy as np
import pandas as pd
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from tqdm.auto import tqdm
import matplotlib.pyplot as plt

# --------------------------------------------------------
# Lightweight ROUGE Implementation (No external library)
# --------------------------------------------------------
def rouge_1(ref, hyp):
    ref_tokens = ref.split()
    hyp_tokens = hyp.split()
    overlap = len(set(ref_tokens) & set(hyp_tokens))
    if len(ref_tokens) == 0: return 0
    return overlap / len(ref_tokens)

def rouge_l(ref, hyp):
    ref_tokens = ref.split()
    hyp_tokens = hyp.split()

    # LCS Calculation
    m, n = len(ref_tokens), len(hyp_tokens)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m):
        for j in range(n):
            if ref_tokens[i] == hyp_tokens[j]:
                dp[i+1][j+1] = dp[i][j] + 1
            else:
                dp[i+1][j+1] = max(dp[i][j+1], dp[i+1][j])

    lcs = dp[m][n]
    if len(ref_tokens) == 0: return 0
    return lcs / len(ref_tokens)


# --------------------------------------------------------
# BLEU Calculation
# --------------------------------------------------------
def calculate_bleu(reference, hypothesis):
    if not reference.strip(): reference = "empty reference"
    if not hypothesis.strip(): hypothesis = "empty hypothesis"

    smoothing = SmoothingFunction().method1

    bleu1 = sentence_bleu(
        [reference.split()],
        hypothesis.split(),
        weights=(1, 0, 0, 0),
        smoothing_function=smoothing
    )

    bleu2 = sentence_bleu(
        [reference.split()],
        hypothesis.split(),
        weights=(0.5, 0.5, 0, 0),
        smoothing_function=smoothing
    )

    return {"bleu1": bleu1, "bleu2": bleu2}


# --------------------------------------------------------
# Prepare Evaluation Sample From Test DF
# --------------------------------------------------------
eval_sample = test_df.sample(n=100, random_state=42)  # 100 samples
print(f"Evaluating on {len(eval_sample)} randomly selected test samples...\n")

our_results = []
base_results = []

for idx, row in tqdm(eval_sample.iterrows(), total=len(eval_sample), desc="Evaluating"):
    context = row["context"]
    reference = row["utterance"]

    # ------------------ OUR CHATBOT (Ollama) ------------------
    our_resp, our_meta = our_chatbot.chat(context)

    bleu_our = calculate_bleu(reference, our_resp)
    rouge1_o = rouge_1(reference, our_resp)
    rougel_o = rouge_l(reference, our_resp)

    our_results.append({
        "bleu1": bleu_our["bleu1"],
        "bleu2": bleu_our["bleu2"],
        "rouge1": rouge1_o,
        "rougeL": rougel_o,
        "empathy": our_meta["response_empathy_score"],
        "sentiment": our_meta["response_sentiment"]
    })

    # ------------------ BASELINE CHATBOT ---------------------
    base_resp, base_meta = baseline_chatbot.chat(context)

    bleu_base = calculate_bleu(reference, base_resp)
    rouge1_b = rouge_1(reference, base_resp)
    rougel_b = rouge_l(reference, base_resp)

    base_results.append({
        "bleu1": bleu_base["bleu1"],
        "bleu2": bleu_base["bleu2"],
        "rouge1": rouge1_b,
        "rougeL": rougel_b,
        "empathy": base_meta["response_empathy_score"],
        "sentiment": base_meta["response_sentiment"]
    })


# --------------------------------------------------------
# Convert to DataFrames
# --------------------------------------------------------
df_our = pd.DataFrame(our_results)
df_base = pd.DataFrame(base_results)

our_avg = df_our.mean()
base_avg = df_base.mean()

comparison = pd.DataFrame({
    "Metric": ["BLEU-1", "BLEU-2", "ROUGE-1", "ROUGE-L",
               "Empathy Score", "Response Sentiment"],

    "Our Chatbot": [
        our_avg["bleu1"], our_avg["bleu2"],
        our_avg["rouge1"], our_avg["rougeL"],
        our_avg["empathy"], our_avg["sentiment"]
    ],

    "Baseline Chatbot": [
        base_avg["bleu1"], base_avg["bleu2"],
        base_avg["rouge1"], base_avg["rougeL"],
        base_avg["empathy"], base_avg["sentiment"]
    ]
})

comparison["Improvement (%)"] = (
    (comparison["Our Chatbot"] - comparison["Baseline Chatbot"])
    / comparison["Baseline Chatbot"].replace(0, np.nan) * 100
)

print("\n" + "="*70)
print("ADVANCED METRIC RESULTS")
print("="*70)
print(comparison.to_string(index=False))


# --------------------------------------------------------
# Visualization
# --------------------------------------------------------
plt.figure(figsize=(14, 7))
x = np.arange(len(comparison["Metric"]))
width = 0.35

plt.bar(x - width/2, comparison["Our Chatbot"], width,
        label="Our Chatbot (Ollama)", color="#2ecc71")

plt.bar(x + width/2, comparison["Baseline Chatbot"], width,
        label="Baseline", color="#e74c3c")

plt.xticks(x, comparison["Metric"], rotation=25, ha="right")
plt.title("Advanced Metrics Comparison")
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig("advanced_metrics_comparison.png", dpi=300)
plt.show()

# --------------------------------------------------------
# Save results
# --------------------------------------------------------
comparison.to_csv("advanced_metrics_summary.csv", index=False)
df_our.to_csv("our_chatbot_detailed_metrics.csv", index=False)
df_base.to_csv("baseline_chatbot_detailed_metrics.csv", index=False)

print("\nâœ“ Advanced Metric Evaluation Completed!")
print("âœ“ Files saved:")
print("   - advanced_metrics_comparison.png")
print("   - advanced_metrics_summary.csv")
print("   - our_chatbot_detailed_metrics.csv")
print("   - baseline_chatbot_detailed_metrics.csv")
```
